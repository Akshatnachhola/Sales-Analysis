```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv ('Diwali Sales Data.csv' , encoding = "ISO-8859-1")
```

```python
df.shape
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] = df['Amount'].astype('int')
```

```python
df['Amount'].dtypes
```

```python
df.columns
```

```python
df.rename(columns = {'Marital_Status' :'Shaadi'})
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

```python
ax = sns.countplot(x='Gender', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Gender',y = 'Amount', data =Sales_gen )
```

```python
df.columns
```

```python
ax = sns.countplot(data= df , x ='Age Group' , hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.barplot(x = 'Age Group',y = 'Amount', data =Sales_Age )
```

```python
Sales_State = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set( rc= {'figure.figsize':(15,5)})
sns.barplot(x = 'State',y = 'Orders', data =Sales_State )
```

```python
Sales_State = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set( rc= {'figure.figsize':(15,5)})
sns.barplot(x = 'State',y = 'Amount', data =Sales_State )
```

```python
ax = sns.countplot(x='Marital_Status', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
ax = sns.countplot(x='Occupation', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Occ = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set( rc= {'figure.figsize':(15,5)})
sns.barplot(x = 'Occupation',y = 'Amount', data =Sales_Occ )
```

```python
ax = sns.countplot(x='Product_Category', data = df)

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Prod= df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.barplot(x = 'Product_Category',y = 'Amount', data =Sales_Prod )
```

```python
Sales_Prod= df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.barplot(x = 'Product_ID',y = 'Orders', data =Sales_Prod )
```
